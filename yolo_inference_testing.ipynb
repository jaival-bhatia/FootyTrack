```python
from ultralytics import YOLO
import torch

model = YOLO("yolov8m.pt")

results = []

for r in model.predict(
    source="input_videos/08fd33_4.mp4",
    stream=True,
    save=True,
    device=0,     # use CUDA
    half=True     # half-precision for performance
):
    #  Step 1: Move to CPU first
    r = r.cpu()

    #  Step 2: Remove large/unneeded components
    r.orig_img = None
    r.masks = None

    torch.cuda.empty_cache()  # optional cleanup

    results.append(r)

# Now safe to access results[0], etc.
print(results[0].boxes.xyxy)  # first frame's bounding boxes
```
