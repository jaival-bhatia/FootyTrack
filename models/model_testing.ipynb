```python
import os
from ultralytics import YOLO
import torch  # optional, but handy if you want to check CUDA availability

# ------------------------------------------------------------------
# 1. File paths (EDIT if your folders move)
# ------------------------------------------------------------------
MODEL_PATH = r"C:\Users\JaivalBhatia\Desktop\All_codes\football_project_opencv\models\best.pt"
VIDEO_PATH = r"C:\Users\JaivalBhatia\Desktop\All_codes\football_project_opencv\input_videos\08fd33_4.mp4"

# ------------------------------------------------------------------
# 2. Quick sanity checks so you fail fast if something’s missing
# ------------------------------------------------------------------
if not os.path.isfile(MODEL_PATH):
    raise FileNotFoundError(f"Model weights not found:\n{MODEL_PATH}")

if not os.path.isfile(VIDEO_PATH):
    raise FileNotFoundError(f"Video file not found:\n{VIDEO_PATH}")

# ------------------------------------------------------------------
# 3. Load model
# ------------------------------------------------------------------
model = YOLO(MODEL_PATH)

# ------------------------------------------------------------------
# 4. Run streamed inference
# ------------------------------------------------------------------
results = []

for r in model.predict(
        source=VIDEO_PATH,
        stream=True,   # yields one result object per frame
        save=True,     # writes an annotated copy next to the input video
        device=0,      # GPU 0 → use 'cpu' if no CUDA
        half=True      # fp16 for extra speed on supported GPUs
):
    results.append(r)
    print(r)  # quick per‑frame overview (classes, conf, etc.)

# ------------------------------------------------------------------
# 5. Post‑processing example
# ------------------------------------------------------------------
if results:
    first_boxes = results[0].boxes.xyxy.cpu().numpy()  # (N, 4) array
    print("\nFirst frame bounding boxes (xyxy):")
    print(first_boxes)
else:
    print("No predictions were made – double‑check video path/format.")
```
